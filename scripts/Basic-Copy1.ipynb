## UMAP overview
UMAP (Uniform Manifold Approximation and Projection) is a nonlinear dimensionality‑reduction method. It builds a k‑nearest‑neighbor graph in the original high‑dimensional space (your flattened EEG features), then learns a 2‑D embedding that best preserves local neighborhood structure. Points that are close in the original space tend to stay close in the UMAP plot, while global distances are only approximate. Results depend on parameters like `n_neighbors` and `min_dist`, and on feature scaling, so the plots are best used for visual pattern discovery rather than strict statistical inference.

```python
# UMAP for each feature sheet, colored by Pain Unpleasantness (T2-T1) median split
import umap
from sklearn.preprocessing import StandardScaler
import numpy as np

output_dir = "../processeddata"

# Get patient IDs with XLSX files
xlsx_files = glob.glob(os.path.join(output_dir, "CIPN3*.xlsx"))
patient_ids = {} #{1: '/rsrch7/home/ip_rsrch/wulab/Mori/EEG/data_txt/all_features/CIPN3001_NFB.xlsx',...}
for f in xlsx_files:
    m = re.search(r"CIPN3(\d{3})", os.path.basename(f))
    if m:
        patient_ids[int(m.group(1))] = f

# Get Pain Unpleasantness differences for these patients
diff_with_xlsx = diff[diff.index.isin(patient_ids.keys())].copy()
median_diff = diff_with_xlsx['Diff'].median()
print(f"Median Pain Unpleasantness (T2-T1): {median_diff:.2f}")
print(f"Patients below median: {(diff_with_xlsx['Diff'] < median_diff).sum()}")
print(f"Patients above median: {(diff_with_xlsx['Diff'] >= median_diff).sum()}\n")

# Define all sheets to process
sheets = [
    "FFT_abs_bandpower_uV2",
    "Z_FFT_abs_bandpower_uV2",
    "FFT_rel_bandpower_pct",
    "Z_FFT_rel_bandpower_pct",
    "FFT_abs_1to50Hz_uV2",
    "Z_FFT_abs_1to30Hz_uV2",
    "Z_FFT_abs_1to50Hz_uV2",
    "FFT_rel_1to50Hz_pct",
    "Z_FFT_rel_1to30Hz_pct",
    "Z_FFT_rel_1to50Hz_pct",
    "PeakFreq_Hz",
    "Z_PeakFreq_Hz",
    "FFT_Coherence",
    "Z_FFT_Coherence",
    "FFT_PhaseLag_PLI",
    "Z_FFT_PhaseLag_PLI"
]

# Process each sheet
for sheet_name in sheets:
    print(f"Processing {sheet_name}...")
    
    # Collect data for all patients
    all_data = []
    patient_nums = []
    colors = []
    
    for patient_num, xlsx_path in patient_ids.items():
        if patient_num not in diff_with_xlsx.index:
            continue
            
        try:
            df_sheet = pd.read_excel(xlsx_path, sheet_name=sheet_name, index_col=0)
            # Flatten the dataframe (channels x features -> 1D vector)
            feature_vector = df_sheet.values.flatten()
            all_data.append(feature_vector)
            patient_nums.append(patient_num)
            
            # Color: below median = blue, above median = red
            if diff_with_xlsx.loc[patient_num, 'Diff'] < median_diff:
                colors.append('blue')
            else:
                colors.append('red')
        except Exception as e:
            print(f"  Error loading {patient_num}: {e}")
            continue
    
    if len(all_data) == 0:
        print(f"  No data found for {sheet_name}\n")
        continue
    
    # Convert to numpy array
    X = np.array(all_data)
    
    # Remove any NaN or Inf values
    X = np.nan_to_num(X, nan=0.0, posinf=0.0, neginf=0.0)
    
    # Standardize
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # UMAP
    reducer = umap.UMAP(n_components=2, random_state=42, n_neighbors=min(5, len(X_scaled)-1))
    embedding = reducer.fit_transform(X_scaled)
    
    # Plot
    plt.figure(figsize=(7, 5))
    for i, color in enumerate(colors):
        plt.scatter(embedding[i, 0], embedding[i, 1], c=color, s=100, alpha=0.6, 
                   edgecolors='black', linewidth=1)
        plt.text(embedding[i, 0], embedding[i, 1], str(patient_nums[i]), 
                fontsize=8, ha='center', va='bottom')
    
    plt.xlabel('UMAP 1', fontsize=12)
    plt.ylabel('UMAP 2', fontsize=12)
    plt.title(f'{sheet_name}\nBlue: < median ({median_diff:.2f}), Red: ≥ median', fontsize=14)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    print(f"  Processed {len(all_data)} patients\n")
```

```python
# UMAP on original (flattened) data after removing values around mean
import matplotlib.pyplot as plt
import umap
import numpy as np
from sklearn.preprocessing import StandardScaler

mean_diff = diff_with_xlsx["Diff"].mean()
mean_margin = 0.5  # adjust as needed

filtered_diff_mean = diff_with_xlsx[abs(diff_with_xlsx["Diff"] - mean_diff) > mean_margin].copy()
filtered_patient_ids_mean = {
    pid: patient_ids[pid]
    for pid in filtered_diff_mean.index
    if pid in patient_ids
}

print(f"Mean Diff: {mean_diff:.2f}")
print(f"Using margin: +/- {mean_margin}")
print(f"Original patients: {len(diff_with_xlsx)}")
print(f"Remaining after filter: {len(filtered_diff_mean)}")

for sheet_name in sheets:
    all_data = []
    ids = []
    for pid, xlsx_path in filtered_patient_ids_mean.items():
        if pid not in filtered_diff_mean.index:
            continue
        df_sheet = pd.read_excel(xlsx_path, sheet_name=sheet_name, index_col=0)
        feature_vector = df_sheet.select_dtypes(include="number").to_numpy().flatten()
        all_data.append(feature_vector)
        ids.append(pid)

    if not all_data:
        print(f"  No data found for {sheet_name}\n")
        continue

    X = np.array(all_data)
    X = np.nan_to_num(X, nan=0.0, posinf=0.0, neginf=0.0)

    X_scaled = StandardScaler().fit_transform(X)
    reducer = umap.UMAP(n_components=2, random_state=42, n_neighbors=min(15, len(X_scaled)-1))
    embedding = reducer.fit_transform(X_scaled)

    median_class = (filtered_diff_mean.loc[ids, "Diff"] >= filtered_diff_mean["Diff"].median()).astype(int)

    plt.figure(figsize=(6, 5))
    plt.scatter(embedding[:, 0], embedding[:, 1], c=median_class, cmap="coolwarm", s=25)
    plt.title(f"UMAP (original): {sheet_name}\nFiltered around mean")
    plt.xlabel("UMAP-1")
    plt.ylabel("UMAP-2")
    plt.colorbar(label="0=below median, 1=above median")
    plt.tight_layout()
    plt.show()
```
